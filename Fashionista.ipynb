# Fashionista  is  a  popular  garment  store  in  Los  Angeles,  USA.  The  store  has multiple  sections  for  men,  women,  and  children  for  different  types  of  garments such  as  trousers,  suits,  traditional  wear,  accessories,  sports  gear,  and  so  on. Currently,  the  store  maintains  a  manual  register  for  recording  the  details  of  the garments  such  as  type,  price,  stock,  orderQty,  sales,  and  so  on.  However,  the owner  of  the  store  has  decided  to  automate  the  tasks  such  as  data  entry,  sales calculation,  stock  updation,  and  so  on  to  save  time  and  effort  required  in  the tasks. 

```python
#Fashionista Assignment
import sys


garments=[] 
sales=0
garmentplst=('trousers', 'suits', 'traditional wear', 'accessories', 'sports gear')

#Adding Garment
def addGarment():
    garment={}                  #To store all garment values we are creating Empty Dic
    garment['gID']=int(input("Enter Garment ID: "))          #garment is we are adding these string inside empty garments which is above
    while True:
        garment['gtype']=input("Enter Garment type-(trousers, suits, traditional wear, accessories, sports gear) :")
        if garment['gtype'] in garmentplst:
            break
        else:
            print("Please enter valied Garment type.....?")
    garment['stock']=int(input("Enter Garment Stock : "))
    garment['price']=int(input("Enter the Garment Price : "))
    garments.append(garment)
    print("Garments added successfully in to the Stock...!")
    print("===============================================")
    
    
    
#Listing Garment
def viewGarment():
    print("\t Displaing all the Garments")
    print("===============================================")
    for item in garments:                    #To fetch every element we are using for loop
        print(str(item['gID']) + "\t" + str(item['gtype']) + "\t" + str(item['stock']) + "\t" + str(item['price']))  #converting from int to string
    print("===============================================")
    
    
#Searching Garment
def searchGarment():
    print("Searching the Garments")
    gID=int(input("Enter Garment ID: "))
    isFound=False                 #Cheking wheter the item is present or not.
    for item in garments:
        if item['gID']==gID:
            print(str(item['gID']) + "\t" + str(item['gtype']) + "\t" + str(item['stock']) + "\t" + str(item['price']))
            isFound=True
    if isFound==False:
        print("Garment Don't Exist with this ID")
    print("===============================================")
    
    
    
#Sale order Calculating
def saleOrder():
    global sales
    print("Calculating Sale Order")
    pid=int(input("Enter Garment ID: "))
    orderqty=int(input("Enter order quality: "))
    flag=True
    for item in garments:
        if pid == item['gID']:
            sales=item['price'] * orderqty
#             garments['stock']=garments['stock'] - orderqty
            flag = False
            break
    if flag:
        print("Invalid Argument ID....!")
    else:
        print("========================================")
        print("Payable Amount is : ", sales )
        print("========================================")
    
    

while True:
    print("================Fashionista Store===============")
    print("\t 1. Add Garment")
    print("\t 2. View Garment")
    print("\t 3. Search Garment")
    print("\t 4. Sale Garment")
    print("\t 5. Exit")
    Choice=int(input("Enter Any Chosse from -(1 to 5):"))
    if Choice == 1:
        addGarment()
        
    elif Choice == 2:
        viewGarment()
        
    elif Choice == 3:
        searchGarment()
        
    elif Choice == 4:
        saleOrder()
        
    elif Choice == 5:
        sys.exit
        
    else:
        print("Invalid Choice, Please try again")
    
```
